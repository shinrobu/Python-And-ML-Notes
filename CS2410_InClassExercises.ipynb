```python
# 3/6/24 Notes
# Identity Matrix stuff continued
def identity_matrix(n):
  result = []
  for row in range(n):
    temp = []
    for col in range(n):
      if row == col:
        temp.append(1)
      else:
        temp.append(0)
    result.append(temp)
  return result

testIdentity = identity_matrix(4)
print(testIdentity)

friendship = [(0,1), (0,2), (1,2), (1,3), (2,3), (3,4), (4,5), (5,6), (5,7), (6,8), (7,8), (8,9)]

def matrix_addition(m1,m2):
  result = []
  if len(m1) != len(m2):
    return None
  else:
    for r in range(len(m1)):
      temp = []
      for c in range(len(m1[0])):
        temp.append(m1[r][c] + m2[r][c])
      result.append(temp)
  return result

m = [[1,3,5,7],[2,4,6,8]]
n = [[2,4,6,8],[1,3,5,7]]
print(matrix_addition(m, n))
# result = [[3,7,11,15], [3,7,11,15]]

def matrix_subtraction(m1,m2):
  result = []
  if len(m1) != len(m2):
    return None
  else:
    for r in range(len(m1)):
      temp = []
      for c in range(len(m1[0])):
        temp.append(m1[r][c] - m2[r][c])
      result.append(temp)
  return result

print(matrix_subtraction(m, n))
```

# 4/10/24: Linear Regression

```python
# 4/10/24: Linear Regression
# Step 1: import libraries
import numpy as np
from sklearn.linear_model import LinearRegression

# Step 2: get data (preprocess/shape as necessary)
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# Step 3: Linear Regression model fitting
model = LinearRegression().fit(x, y)

# Step 4a: get result (slope and intercept)
print("intercept: ", model.intercept_)
print("slope: ", model.coef_)

# Step 4b: validate
r_sq = model.score(x, y)
print(f"coefficient of determination: {r_sq}")

# Step 5: use model to predict values
y_pred = model.predict(x)
print(f"predicted response:\n{y_pred}")

# Step 5 cont: use model to predict with new values
x_new = np.arange(5).reshape((-1,1))
y_new = model.predict(x_new)
print('predicted response for new x values: ', y_new)
```
